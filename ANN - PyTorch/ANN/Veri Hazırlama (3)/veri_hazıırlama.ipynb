# 1Ô∏è‚É£ üîß Gerekli K√ºt√ºphaneler ve Importlar

```python
# 1Ô∏è‚É£ üîß Gerekli K√ºt√ºphaneler ve Importlar
import torch
from torch.utils.data import Dataset, DataLoader
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
import pandas as pd
import numpy as np
```

# 2Ô∏è‚É£ üìÇ Veri Y√ºkleme

```python
df = pd.read_csv(r"C:\Users\hdgn5\OneDrive\Masa√ºst√º\pytorch fundementals\ANN\Datasets\cancer.csv")
df.head()
```

```python
df.isna().sum()
```

```python
df = df.drop(columns=["Unnamed: 32"])
df.head()
```

# 3Ô∏è‚É£ üéØ √ñzellik ve Hedef Ayƒ±rma

```python

x = df.drop("diagnosis" ,axis=1).values
y = df["diagnosis"].values

le = LabelEncoder()
y = le.fit_transform(y)  # "M"/"B" ‚Üí 1/0
```

# 4Ô∏è‚É£ üìù Eƒüitim/Test Seti B√∂lme

```python

X_train, X_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
) 
```

# 5Ô∏è‚É£ ‚öñÔ∏è √ñzellikleri Normalle≈ütirme

```python

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)  # üîπ Eƒüitim verisini √∂ƒürenip d√∂n√º≈üt√ºr
X_test = scaler.transform(X_test)        # üîπ Test verisini aynƒ± √∂l√ßekle d√∂n√º≈üt√ºr
```

# 6Ô∏è‚É£ üî¢ Tensorlara √áevirme

```python

import torch

X_train = torch.tensor(X_train, dtype=torch.float32)
X_test = torch.tensor(X_test, dtype=torch.float32)
y_train = torch.tensor(y_train, dtype=torch.float32).view(-1, 1)
y_test = torch.tensor(y_test, dtype=torch.float32).view(-1, 1)
```

# 7Ô∏è‚É£ üì¶ Dataset ve DataLoader Olu≈üturma

 # üì¶ DataLoader Nedir?

`DataLoader`, PyTorch‚Äôta **Dataset‚Äôten veri √ßekmek ve modelin eƒüitimine hazƒ±r h√¢le getirmek i√ßin kullanƒ±lan bir sƒ±nƒ±ftƒ±r**.

### Temel ƒ∞≈ülevleri

1. **Mini-batch‚Äôler olu≈üturur**
   - B√ºy√ºk veri k√ºmelerini k√º√ß√ºk par√ßalar halinde modele verir.
   - Hafƒ±za tasarrufu ve daha stabil √∂ƒürenme saƒülar.

2. **Veriyi karƒ±≈ütƒ±rabilir (shuffle)**
   - `shuffle=True` ile her epoch ba≈üƒ±nda veriyi karƒ±≈ütƒ±rƒ±r.

3. **Multi-thread ile veri y√ºkler (num_workers)**
   - B√ºy√ºk veri k√ºmelerinde veriyi arka planda hƒ±zlƒ±ca hazƒ±rlar.


### Kullanƒ±mƒ±

- √ñnce bir **Dataset** sƒ±nƒ±fƒ± olu≈üturulur.
- Dataset, `__len__` ve `__getitem__` metodlarƒ±nƒ± implement eder.
- Dataset‚Äôi `DataLoader`‚Äôa veririz:

# üìÇ Dataset Nedir?

`Dataset`, PyTorch‚Äôta **veri kaynaƒüƒ±nƒ± temsil eden sƒ±nƒ±ftƒ±r**.  

- Veri k√ºmesini **tek bir nesne altƒ±nda toplar**.  
- `DataLoader` ile birlikte kullanƒ±ldƒ±ƒüƒ±nda veriyi modelin anlayacaƒüƒ± ≈üekilde hazƒ±rlar.  

## Temel G√∂revleri

1. **Veri Saklama**  
   - T√ºm √∂zellikler (X) ve hedefler (y) Dataset i√ßinde tutulur.  

2. **Veriye Eri≈üim Saƒülama**  
   - `__getitem__(self, idx)` ‚Üí ƒ∞ndex numarasƒ±na g√∂re tek bir √∂rneƒüi verir.  

3. **Veri Uzunluƒüu Saƒülama**  
   - `__len__(self)` ‚Üí Dataset‚Äôin toplam √∂rnek sayƒ±sƒ±nƒ± verir.  

## Kodun yapƒ±sƒ±na ge√ßmeden √∂nce size __len__ ve __getitem__ fonksiyonlarƒ±nƒ±n ne i≈üe yaradƒ±ƒüƒ±nƒ± anlatayƒ±m.Burada kullanƒ±lacak Dataset ve Loaderlarƒ±n i≈ülevselliƒüinin daha geni≈ü olmasƒ± i√ßin a≈üaƒüƒ±daki bilgileri bƒ±rakƒ±yorum

* __len__ ‚Üí Toplam √∂rnek sayƒ±sƒ±nƒ± d√∂nd√ºr√ºr

* __getitem__ ‚Üí ƒ∞lgili indeksteki √∂rneƒüi d√∂nd√ºr√ºr

`x ‚Üí Modelin giri≈ü verisi`

`y ‚Üí Modelin tahmin edeceƒüi √ßƒ±kƒ±≈ü (target)`

* Yani Dataset ve DataLoader olu≈ütururken de bu x ve y tensorlarƒ±nƒ± kullanƒ±yoruz.

### üìÇ Dataset

```python
from torch.utils.data import Dataset

class MyDataset(Dataset):
    def __init__(self, X, y):
        self.X = X
        self.y = y
    
    def __len__(self):
        return self.X.shape[0]   # len yerine shape[0] kullanmak daha g√ºvenli
    
    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]
```

### üì¶ DataLoader

```python
from torch.utils.data import DataLoader

train_dataset = MyDataset(X_train, y_train)
test_dataset = MyDataset(X_test, y_test)

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)
```
